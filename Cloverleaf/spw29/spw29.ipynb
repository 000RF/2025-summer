### binning + boxcar

```python
def boxcar_smooth(y, kernel_size):
    kernel = np.ones(kernel_size)/kernel_size
    return np.convolve(y, kernel, mode="same")

def rebin_spectrum(x, y, bin_size):
    x = np.array(x)
    y = np.array(y)
    n_bins = len(y) // bin_size
    x = x[:n_bins * bin_size]
    y = y[:n_bins * bin_size]
    x_binned = x.reshape(n_bins, bin_size).mean(axis=1)
    y_binned = y.reshape(n_bins, bin_size).mean(axis=1)
    return x_binned, y_binned
```

```python
x_binned, y_binned = rebin_spectrum(x, y, bin_size=2)

y_binned_boxcar = boxcar_smooth(y_binned, kernel_size=6)


plt.figure(figsize=(8, 4))
plt.plot(x_binned, y_binned_boxcar, color="black", linewidth=1, label="binning(b=2)+Boxcar (k=6)")
for freq, label in my_lines:
    plt.axvline(freq, color="red", linestyle="-", alpha=0.6)
    plt.text(freq, max(y_binned_boxcar)*-0.05, label, rotation=90, color="red", fontsize=10, ha='right', va='top')
plt.xlabel("frequency (GHz)")
plt.ylabel("Flux Density (mJy/beam)")
plt.title("Cloverleaf spw29")
plt.xlim(108.25, 110)
plt.ylim(-0.20, 0.30)
plt.axhline(0, color="black", linewidth=1, linestyle="--")
plt.grid(True, alpha=0.2)
plt.legend()
```
