```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
img1 = mpimg.imread("spw25(L).png")
img2 = mpimg.imread("spw27(L).png")
img3 = mpimg.imread("spw29(L).png")
img4 = mpimg.imread("spw31(L).png")
```

```python
fig, axs = plt.subplots(2, 2, figsize=(20, 10)) 
axs[0, 0].imshow(img1)
axs[0, 0].set_title('spw25')
axs[0, 0].axis('off')

axs[0, 1].imshow(img2)
axs[0, 1].set_title('spw27')
axs[0, 1].axis('off')

axs[1, 0].imshow(img3)
axs[1, 0].set_title('spw29')
axs[1, 0].axis('off')

axs[1, 1].imshow(img4)
axs[1, 1].set_title('spw31')
axs[1, 1].axis('off')
```
