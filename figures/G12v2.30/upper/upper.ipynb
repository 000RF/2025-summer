```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
img1 = mpimg.imread("spw21.png")
img2 = mpimg.imread("spw23.png")
img3 = mpimg.imread("spw25.png")
img4 = mpimg.imread("spw27.png")
```

```python
fig, axs = plt.subplots(2, 2, figsize=(20, 10)) 
axs[0, 0].imshow(img1)
axs[0, 0].set_title('spw21')
axs[0, 0].axis('off')

axs[0, 1].imshow(img2)
axs[0, 1].set_title('spw23')
axs[0, 1].axis('off')

axs[1, 0].imshow(img3)
axs[1, 0].set_title('spw25')
axs[1, 0].axis('off')

axs[1, 1].imshow(img4)
axs[1, 1].set_title('spw27')
axs[1, 1].axis('off')

fig.suptitle("G12v2.30(all)", fontsize=20)
```
