```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
from mpl_toolkits.axes_grid1.inset_locator import mark_inset
from matplotlib.patches import ConnectionPatch
```

```python
img1 = mpimg.imread("spw25.png")
img2 = mpimg.imread("spw27.png")
img3 = mpimg.imread("spw27(CN).png")
img4 = mpimg.imread("spw29.png")
img5 = mpimg.imread("spw31.png")
```

```python
fig, axs = plt.subplots(2, 2, figsize=(20, 10)) 
axs[0, 0].imshow(img1)
axs[0, 0].set_title('spw25')
axs[0, 0].axis('off')

axs[0, 1].imshow(img2)
axs[0, 1].set_title('spw27')
axs[0, 1].axis('off')

axins = inset_axes(
    axs[0, 1],
    width="50%", height="50%",
    loc="upper center",
    bbox_to_anchor=(0, -0.2, 1.1, 1),
    bbox_transform=axs[0, 1].transAxes,
    borderpad=0
)
axins.imshow(img3)
x1, x2 = 0, 683
y1, y2 = 0, 351
axins.set_xlim(x1, x2)
axins.set_ylim(y2, y1)
axins.axis('off')

con = ConnectionPatch(
    xyA=(190, 300), coordsA=axs[0, 1].transData,  
    xyB=(0.08, 0.0), coordsB=axins.transAxes,
    axesA=axs[0, 1], axesB=axins,
    color="red", lw=2
)
fig.add_artist(con)

axs[1, 0].imshow(img4)
axs[1, 0].set_title('spw29')
axs[1, 0].axis('off')

axs[1, 1].imshow(img5)
axs[1, 1].set_title('spw31')
axs[1, 1].axis('off')
```
